```python
import requests
from bs4 import BeautifulSoup

def crawl_halodoc(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    paragraphs = soup.find_all("p")
    text = " ".join([p.get_text() for p in paragraphs])
    return text

def crawl_alodokter(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    article = soup.find("div", {"class": "content-article"})
    if article:
        paragraphs = article.find_all("p")
        text = " ".join([p.get_text() for p in paragraphs])
        return text
    return ""

all_urls = [
    "https://www.halodoc.com/artikel/kenali-gejala-dan-cara-mengatasi-flu-biasa",
    "https://www.halodoc.com/artikel/penyebab-dan-cara-mengatasi-sakit-tenggorokan",
    "https://www.halodoc.com/artikel/benarkah-stres-bisa-mengganggu-sistem-imun",
    "https://www.halodoc.com/artikel/tanda-awal-diabetes-yang-perlu-diwaspadai",
    "https://www.halodoc.com/artikel/cara-menjaga-kesehatan-jantung",
    "https://www.halodoc.com/artikel/7-cara-meningkatkan-kualitas-tidur",
    "https://www.halodoc.com/artikel/cara-menurunkan-tekanan-darah-tinggi",
    "https://www.halodoc.com/artikel/penyebab-kolesterol-tinggi",
    "https://www.halodoc.com/artikel/penyebab-sakit-kepala-berulang",
    "https://www.halodoc.com/artikel/cara-menjaga-kesehatan-mental",
    "https://www.alodokter.com/gejala-flu-yang-perlu-diketahui",
    "https://www.alodokter.com/penyebab-sakit-tenggorokan-dan-cara-mengatasinya",
    "https://www.alodokter.com/penjelasan-tentang-diabetes",
    "https://www.alodokter.com/kolesterol-tinggi",
    "https://www.alodokter.com/tekanan-darah-tinggi",
    "https://www.alodokter.com/kenali-depresi-dan-gejalanya",
    "https://www.alodokter.com/cara-menjaga-kesehatan-jantung",
    "https://www.alodokter.com/penyakit-asma",
    "https://www.alodokter.com/hipertensi",
    "https://www.alodokter.com/cara-meningkatkan-imunitas-tubuh"
    ]

docs = []
for url in all_urls:
    try:
        if "halodoc" in url:
            text = crawl_halodoc(url)
        else:
            text = crawl_alodokter(url)

        docs.append(text)
        print(f"Berhasil: {url}")
    except:
        print(f"Gagal: {url}")
```
